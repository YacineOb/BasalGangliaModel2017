```python
wrilist = []
Gbilist = []

for wi in tqdm(np.arange(0.001, 100, 0.001)):
    Gbi = (1 - ((4*wi*tc)*((wi*tc)**2 - 1) / sin(wi*DeltaA))) / (1-wi*tc*(1/tan(wi*DeltaA)))
    Gbilist.append(Gbi)
    wrilist.append(wi)
        

plt.plot(Gbilist, wrilist)
plt.xlabel('Gb')
plt.ylabel('w')
plt.title('with Ga fixed to 1')

plt.show()
```
